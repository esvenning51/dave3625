```python
import pandas as panda
from matplotlib import pyplot as plt
```

```python
dataframe = panda.read_csv("ruter_data.csv",sep=';')
```

```python
dataframe.shape
```

```python
dataframe.head()
```

```python
plt.hist(dataframe.Kjøretøy_Kapasitet, bins=10, rwidth=1)
plt.xlabel('Vehicle Capacities')
plt.ylabel('Count')
plt.show()
```

```python
dataframe.plot.scatter(x='Kjøretøy_Kapasitet', y='Passasjerer_Ombord', title='Usage vs Capacity')
```

<strong>^Has to have some wrong values because of negative amount of passengers for some of the trips. </strong>

```python
dataframe = dataframe[dataframe['Passasjerer_Ombord'] >= 0]
dataframe
```

<strong>Negative values for passengers removed:</strong>

```python
dataframe.plot.scatter(x='Kjøretøy_Kapasitet', y='Passasjerer_Ombord', title='Capacity vs. Usage')
```

<h4>Trips where more than 40% of the capacity is used:</h4>

```python
over_40pct = dataframe[dataframe['Kjøretøy_Kapasitet'] - dataframe['Passasjerer_Ombord'] < dataframe['Kjøretøy_Kapasitet']*0.6]
over_40pct
```

```python
over_40pct["Tidspunkt_Planlagt_Avgang_Holdeplass_Fra"]
```

<h4>Time of day for these trips (over 40% of capacity used):</h4>

```python
time40pct = panda.to_datetime(over_40pct['Tidspunkt_Planlagt_Avgang_Holdeplass_Fra']).dt.time
time40pct
```

```python
time40pct = panda.to_datetime(over_40pct['Tidspunkt_Planlagt_Avgang_Holdeplass_Fra']).dt.hour
time40pct
```

```python
time40pct.value_counts().sort_index().plot.bar(width=0.8,color='g')
```

<h2> Line number frequency for capacity used >40% :</h2>

```python
over_40pct["Linjenavn"].value_counts().plot.barh(color='r')
```

```python
line60 = over_40pct[(over_40pct["Linjenavn"] == "60")]
line60
```

<h4>Direction of line 60 when above 40% capacity used:</h4>

```python
line60["Linjeretning"].value_counts().plot.bar(color='y')
plt.xticks(rotation = 0)
```
